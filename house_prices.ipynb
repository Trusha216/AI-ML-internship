```python
data.rename(columns={"Carpet Area": "Carpet Area (in sqft)"}, inplace=True)
```

```python
data.rename(columns={"Super Area": "Super Area (in sqft)"}, inplace=True)
```

```python
data["Carpet Area (in sqft)"] = data["Carpet Area (in sqft)"].str.replace(" sqft", "", regex=True)
data["Carpet Area (in sqft)"] = pd.to_numeric(data["Carpet Area (in sqft)"], errors="coerce")

# Display updated DataFrame
print(data.head())
```

```python
data["Super Area (in sqft)"] = data["Super Area (in sqft)"].astype(str).str.replace(" sqft", "", regex=True)
data["Super Area (in sqft)"] = pd.to_numeric(data["Super Area (in sqft)"], errors="coerce")

print(data.head())  # Check the updated DataFrame
```

```python
data["Open car Parking"] = data["Car Parking"].str.extract(r'(\d+)\s*Open')

data["Covered car Parking"] = data["Car Parking"].str.extract(r'(\d+)\s*Covered')

data["Open car Parking"] = pd.to_numeric(data["Open car Parking"], errors="coerce").fillna(0)
data["Covered car Parking"] = pd.to_numeric(data["Covered car Parking"], errors="coerce").fillna(0)
```

```python
data.head(10)
```

```python
import pandas as pd
import numpy as np
floor_data = data['Floor']

current_floor = []
total_floors = []

for entry in floor_data:
    if isinstance(entry, str) and " out of " in entry:
        parts = entry.split(" out of ")
        floor = parts[0].strip()
        total = parts[1].strip()


        if floor.lower() == "ground":
            current_floor.append(0)
        elif floor.lower() == "upper basement":
            current_floor.append(-1)
        elif floor.lower() == "lower basement":
            current_floor.append(-2)
        else:
            try:
                current_floor.append(int(floor))
            except ValueError:
                current_floor.append(np.nan)


        try:
            total_floors.append(int(total))
        except ValueError:
            total_floors.append(np.nan)
    else:

        current_floor.append(np.nan)
        total_floors.append(np.nan)


df = pd.DataFrame({"Current Floor": current_floor, "Total Floors": total_floors})


print(df)
```

```python
df = pd.DataFrame({"Current Floor": current_floor, "Total Floors": total_floors})
df.to_csv("house_prices.csv", mode='a', index=False, header=False)
```

```python
data["Current Floor"] = current_floor
data["Total Floors"] = total_floors

data.to_csv("house_prices.csv", index=False)
```

```python
def convert_cr_to_lac(amount):
    if isinstance(amount, str):
        amount = amount.strip()
        if "Cr" in amount:
            value = float(amount.replace("Cr", "").strip())
            return f"{value * 100} Lac"
        elif "Lac" in amount:
            return amount  # Already in Lac
        elif "Call for Price" in amount:
            return amount  # Keep as is
    return "Invalid Input"

# Example Usage:
amounts =data['Amount(in lac)']
converted_amounts = [convert_cr_to_lac(a) for a in amounts]
```

```python
data["converted_amounts"] = converted_amounts
data.to_csv("house_prices.csv", index=False)
```
